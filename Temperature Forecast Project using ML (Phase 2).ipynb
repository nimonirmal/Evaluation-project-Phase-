# Project Description
This data is for the purpose of bias correction of next-day maximum and minimum air temperatures forecast of the LDAPS model operated by the Korea Meteorological Administration over Seoul, South Korea. This data consists of summer data from 2013 to 2017. The input data is largely composed of the LDAPS model's next-day forecast data, in-situ maximum and minimum temperatures of present-day, and geographic auxiliary variables. There are two outputs (i.e. next-day maximum and minimum air temperatures) in this data. Hindcast validation was conducted for the period from 2015 to 2017.

# <span style="color:purple">Checking Outliers

```python
#checking outliers by using Boxplot
plt.figure(figsize=(20,25))
p=1
for i in df:
    if p<=21:
        plt.subplot(7,3,p)
        sns.boxplot(y=df[i], palette="Set2_r")
        plt.xlabel(i,fontsize =20)
    p+=1
plt.show()
```

<span style="color:blue">Observation- Some Features has Outliers.Lets remove Outliers by IQR method or Z-score method.

```python
#Remove Outliers by Z-score method
from scipy.stats import zscore
outliers_feature = df[['Present_Tmax', 'Present_Tmin','LDAPS_RHmax','LDAPS_Tmax_lapse', 'LDAPS_Tmin_lapse',
                       'LDAPS_WS', 'LDAPS_LH','DEM', 'Slope']]

z=np.abs(zscore(outliers_feature)) #.abs() to remove negative values
```

```python
#now removing the data above zscore and creating a new dataframe
df1=df[(z<3).all(axis=1)]
print("New Dataframe having rows and column: " ,df1.shape)
print("New Dataframe column's name: " ,df1.columns)
```

```python
print("Data Loss Percentage with Z-score method:",((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

```python
#Remove Outliers by IQR method
#1st Quantile
q1=outliers_feature.quantile(0.25)

#3rd Quantile
q3=outliers_feature.quantile(0.75)

#calculating IQR
iqr=q3-q1

df2=df[~((outliers_feature < (q1 - 1.5*iqr ))|(outliers_feature > (q3 + 1.5*iqr ))).any(axis=1)]
```

```python
print("Data Loss Percentage with Z-score method:",((df.shape[0]-df2.shape[0])/df.shape[0])*100)
```

<span style="color:blue">Conclusion -Data Loss Percentage  with IQR method is very high as compared with Z-score . So i go with z-score.

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

def model_score(model_name,x_train,x_test,y_train,y_test, train=True):
    if train: #checking training score 
        y_pred_train=model_name.predict(x_train)
        
        print("\n*********************Train Result*********************")
        
        print("R2 Score on train data:",r2_score(y_train,y_pred_train))
        
    elif train==False: #checking testing score 
        y_pred_test=model_name.predict(x_test)
        
        print("\n*********************Test Result*********************")
        
        print("R2 Score on test data:",r2_score(y_test,y_pred_test))
        print("Mean Absolute Error :",mean_absolute_error(y_test,y_pred_test))
        print("Mean Squared Error :",mean_squared_error(y_test,y_pred_test))
        print("Root Mean Squared Error :",np.sqrt(mean_squared_error(y_test,y_pred_test))) 
        
        print("\n*********************Cross-Validation score*********************")
        print("Overall Cross-Validation score :",cross_val_score(model_name,x_scaled, y,cv=5).mean())
        
        print("\n************Difference in CV score and Accuracy score***********")
        deviation= ((r2_score(y_test,y_pred_test))-(cross_val_score(model_name,x_scaled, y,cv=5).mean()))
        print("Difference",deviation)
```
